# Part 3: Exploratory Analysis

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':{'$eq':20}}
fields ={'AddressLine1':1,'AddressLine2':1,'AddressLine3':1,'BusinessName':1,'BusinessType':1,'BusinessTypeID':1,
         'ChangesByServerID':1,'Distance':1,'FHRSID':1,'LocalAuthorityBusinessID':1,'LocalAuthorityCode':1,'LocalAuthorityCode':1,
         'LocalAuthorityEmailAddress':1,'LocalAuthorityName':1,'LocalAuthorityWebSite':1,'PostCode':1,'RatingDate':1,'RatingValue':1,
         'geocode':1, 'scores':1}
         
hygien_score = list(establishments.find(query,fields))
# Use count_documents to display the number of documents in the result

pprint(f'total number of hygien {establishments.count_documents(query)}')
print(f'____________________________________________________________________')

# Display the first document in the results using pprint

hygien_score = list(establishments.find(query))


pprint(hygien_score[0])




```

```python
# we need to use json_normalize for the nest dictionary
# Convert the result to a Pandas DataFrame
df = pd.json_normalize(hygien_score)


df.head(10)
```

```python
# Display the number of rows in the DataFrame
pprint(df.shape)
```

```python
#Which establishments in London have a RatingValue greater than or equal to 4?
query = {'LocalAuthorityName':{'$regex':"London"} ,'RatingValue':{'$gte':4}} #'LocalAuthorityName'
field = {'RatingValue':1,'LocalAuthorityName':1}

# Display the first document in the results using pprint
result = establishments.find(query,field)
pprint(result[0])
```

```python
# Use count_documents to display the number of documents in the result
establishments.count_documents(query)
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(result)
```

```python
# Display the number of rows in the DataFrame
pprint(london_df.shape)
```

 # what are the top 5 establishments with a rating value of 5,sorted by the lowest hygiene score.nearest to the new restaurant added " Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 0.08384000   # latitude for "Penang Flavours
longitude =51.49014200  # longitude for new restuarnt "Penang Flavours.
# our search will between 0.01+latitude and 0.01+longitude

query ={'geocode.latitude':{'$gte':0.09384},'geocode.longitude':{'$lte':51.500142,},'RatingValue':{'$eq':5}}
sort =[('scores.Hygiene',1)]
limit =5
results = list(establishments.find(query).sort(sort).limit(limit))
# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
top_five_df = pd.json_normalize(results)
top_five_df
```

# how many establishment in each Local Authority area have a hygiene score of 0

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
hygien_score = {'$match':{'scores.Hygiene':{'$eq':0}}}
# 2. Groups the matches by Local Authority
group = [{'$group':{'_id':'$LocalAuthorityName','count':{'$sum':1}}}]
# 3. Sorts the matches from highest to lowest
#sorts = {'$sort':{'count',-1}}
# Print the number of documents in the result
pipe_line = [hygien_score,group]
result = list(establishments.aggregate(hygien_score,group))
# Print the first 10 results
pprint(result[0:10])
```
